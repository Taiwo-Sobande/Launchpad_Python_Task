2. Data Cleaning

```python
# Convert price fields into float

df_clean= df.copy()
df_clean['price'] = pd.to_numeric(df_clean['price'], errors ='coerce')
```

```python
# Parse last_review field to a datetime object

df_clean['last_review']= pd.to_datetime(df_clean['last_review'],errors= 'coerce')
```

```python
# Handle missing values

df_clean['reviews_per_month'] = df_clean['reviews_per_month'].fillna(0)
df_clean['host_name']= df_clean['host_name'].fillna('Unknown')
df_clean['neighbourhood_group']= df_clean['neighbourhood_group'].fillna('Unknown')

df_clean.isnull().sum()
```

```python
# Remove irrelevant rows (Listings with zero availability or price <=0 are irrelevant for analysis by the customer for comparison)

df_clean = df_clean[
                    (~df_clean['availability_365'].isna())
                     & (df_clean['availability_365']> 0)
                     & (~df_clean['price'].isna())
                     & (df_clean['price']> 0)
]

df_clean= df_clean.drop_duplicates()
```

3. Data Enrichment

```python
# Create a price_per_booking column using price and minimum_nights.

df_clean['price_per_booking'] = df_clean['price']* df_clean['minimum_nights']
```

```python
# Bucket availability into categories

def bucket(x):
    if pd.isna(x):
        return 'Unknown'
    if x < 100:
        return 'Rare'
    if 100<= x <=300:
        return 'Part-time'
    if x>300:
        return 'Full-time'

df_clean['availability_category']= df_clean['availability_365'].apply(bucket)
```

4. Data Analysis Using Pandas

```python
df_clean.info()
```

```python
#What are the top 10 most expensive neighborhoods by average price?
top_10 = (df_clean.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10).reset_index().rename(columns={'price':'average_price'}))

top_10
```

```python
# What’s the average availability and price by room type?

avg_by_room_type= (df_clean.groupby('room_type').agg(average_availability=('availability_365', 'mean'), average_price=('price','mean')).reset_index().sort_values('average_price', ascending= False))

avg_by_room_type
```

```python
# Which host has the most listings?

#top_hosts= (df_clean.groupby('host_name').size().reset_index(name= 'm_listing').sort_values('m_listing', ascending= False).head(5))

top_hosts = (df_clean[df_clean['host_name']!='Unknown'].groupby('host_name')['calculated_host_listings_count'].sum().sort_values(ascending=False).head(10).reset_index().rename(columns={'calculated_host_listings_count':'listings_count'}))

top_hosts
```

```python
#How does average price vary across different boroughs or districts?

average_price_borough = (df_clean.groupby('neighbourhood_group')['price'].mean().reset_index().sort_values('price', ascending= False).rename(columns={'price':'average_price'}))

average_price_borough
```

```python
# How many listings have never been reviewed?

not_reviewed= (df_clean['number_of_reviews']== 0).sum()

print(f"\nCount of Listings never reviewed: {not_reviewed}")
```
